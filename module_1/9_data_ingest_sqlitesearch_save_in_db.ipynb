```python
from ingest import load_faq_data
from rag_helper import RAGBase
from sqlitesearch import TextSearchIndex

# load the documents
documents = load_faq_data()
# index = build_index(documents) we will not use misearch instead
# we will select only the llm-zoomcamp since we will save the data in sqilte database and we can choose any DB
docs_llm = [doc for doc in documents if doc['course'] == 'llm-zoomcamp']
print(len(docs_llm))

# set the index
sqlite_index = TextSearchIndex(
    text_fields=['question', 'section', 'answer'],
    # keyword_fiels=['course'],
    db_path='feq.db' # set the db name
)

```

# ingest the data (documents) to DB

```python
import time

for doc in docs_llm:
    sqlite_index.add(doc)
    print(f'Added: {doc["question"][:60]}.....')
    time.sleep(0.5)
```

```python
# testing only
sqlite_index.search("how can I join the course")
```
