```python
#########################################################################################################   
# Instead of using openAI, we can either use Huggingface on the cloud with free key, or use ollama server
########################################################################################################## 
# 1.  Use langchain_ollma
#-------------------------------------------------------------------------
# 1.1 download ollama server:: https://ollama.com/download
# 1.2. open a powershell: write olla --version, then start ollama server:  ollama serve
# 1.3. check it, open a new powershell and write:   curl.exe http://localhost:11434 and  press Y
# 1.4. pull the modle:: ollama pull gemma3:4b
# 1.5. run the server:  ollama run gemma3:4b

from langchain_ollama import ChatOllama

question = "what is langchain"
llm_ollama = ChatOllama(
    model="gemma3:4b",
    base_url="http://localhost:11434",
    temperature = 0,
)
llm_ollama_response = llm_ollama.invoke(question)
print(f"Answer fom llm_ollama is::: {llm_ollama_response.content}")
```

```python
# 2.  Use HuggingFace
#-------------------------------------------------------------------------
from huggingface_hub import InferenceClient

# Use the exact model and token that you verified is working!
client = InferenceClient(
    model="Qwen/Qwen2.5-Coder-32B-Instruct",
    token=os.getenv("HUGGINGFACE_DATA_TALK_CAMP_TOKEN"), 
)

response = client.chat_completion(
        messages=[
            {"role": "user",
             "content": question}
        ],
        # max_tokens=100,
        temperature=0,
    )
print(response.choices[0].message.content)
# messages=[
#     {
#         "role": "user", 
#         "content": """
#         Answer the question based strictly on the context provided below.
        
#         Context: [Your Python script will dynamically insert a PDF paragraph here]
        
#         Question: What is the maintenance procedure for machine X?
#         """
#     }
# ]
```
