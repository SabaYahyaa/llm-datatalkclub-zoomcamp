# Create ground truth data to evaluate RAG response

source: https://www.youtube.com/watch?v=eC_IcxfxoiQ&list=PL3MmuxUbc_hLZFNgSad56pDBKK8KO0XIv



There are 2 ways to get data:
1. give question to RAG and then collect the result, and we check the logs (this is tking a lot of time)
2. another way is to take the answer A, and wa generate several questions that have the answer, then we give the question to RAG and we check if the anwser from RAG matches the correct answer A.

### The evaluation setup (2nd way) [Search evaluation]
-  A = the original answer from our data (FAQ)

-  Q* = a question generated from the answer A by an LLM

- we send Q* through our search and we check if the original document appears in the results (if the answer appears in the result)

### For RAG evaluation

-  A = the original answer from our data (FAQ)

-  Q* = a question generated from the answer A by an LLM

-  A' = the answer produced by our RAG system when given Q*

- We compare A' with A to see if the system produced the right answer


This is the A → Q* → A' pattern. We know the answer for each generated question because we created the question from that answer.

- With evaluation, we can:

Compare different search methods (minsearch vs vector search vs hybrid)

Tune parameters (boost values, number of results, prompt templates)

Compare different LLMs (gpt-5.4-mini vs others)

Track improvements over time

### Types of Evaluation

- offline: run the system on a test dataset and compute metrics (Offline evaluation is what we do before putting changes in front of users. It lets us compare search settings, prompts, or models on the same dataset. )
- online: collect feedback from real users in production (Online evaluation happens after deployment. It uses real traffic, feedback, logs, and dashboards to monitor quality.)


Synthetic data is a good starting point when you don't have real user data. But generated questions can be too similar to the original FAQ text, which inflates the metrics. As soon as you can, start collecting real user queries and use them to validate your evaluation framework.

We'll cover three levels of evaluation:

Search evaluation: does the search return the right documents?

RAG evaluation: does the LLM generate good answers?

Agent evaluation: does the agent use tools efficiently?


Most of our time goes to search, and that's on purpose. 

Everything else depends on it: if retrieval brings back the wrong documents, no prompt or model can rescue the answer. So we test search on its own first, then evaluate the full pipeline on top of it.




For search, we'll use two metrics: 

- Hit Rate and 

- MRR (Mean Reciprocal Rank). 


For RAG quality, we'll use 

- LLM-as-a-judge. For agents, we'll look at the final answer and the tool-call trajectory.
